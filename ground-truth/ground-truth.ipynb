```python
from sympy import Point, Line2D
import json
import pandas as pd
```

```python
with open('points.json') as json_data:
    data = json.load(json_data)
    json_data.close()
    print(data)
```

```python
print(data[0]['file_upload'])

        
```

```python
df = pd.DataFrame(columns = ["file_name", "left_radius", "right_radius"])
```

```python
for i, item in enumerate(data):
    right_radius = -1
    left_radius = -1
    file_name = item["file_upload"]

    cm = item["annotations"][0]["result"][-1]["value"]["points"]
    dist = abs(cm[0][0] - cm[1][0])

    print(dist)
    for label in item["annotations"][0]["result"]:
        if label["value"]["polygonlabels"][0] == "right-poly":

            right_poly = label["value"]["points"]


            ab = sorted(sorted(right_poly, key=lambda x: x[0])[:2], key=lambda x: -x[1])

            cd = sorted(sorted(right_poly, key=lambda x: -x[0])[:2], key=lambda x: x[1])

            bc = [ab[1], cd[0]]

            ad = [ab[0], cd[1]]

            # plt.scatter(ab[0][0], -ab[0][1], color = "blue")
            # plt.scatter(ab[1][0], -ab[1][1], color = "red")
            # plt.scatter(cd[0][0], -cd[0][1], color = "purple")
            # plt.scatter(cd[1][0], -cd[1][1], color = "orange")
            a = Point(ab[0][0], ab[0][1])

            ab = Line2D(Point(ab[0][0], ab[0][1]), Point(ab[1][0], ab[1][1] - 0.5 * dist))
            cd = Line2D(Point(cd[0][0], cd[0][1] - 0.5 * dist), Point(cd[1][0], cd[1][1]))
            bc = Line2D(Point(bc[0][0], bc[0][1] - 0.5 * dist), Point(bc[1][0], bc[1][1] - 0.5 * dist))
            ad = Line2D(Point(ad[0][0], ad[0][1]), Point(ad[1][0], ad[1][1]))

            bis1 = bc.bisectors(ab)[1] 
            bis2 = bc.bisectors(cd)[1]
            center = bis1.intersection(bis2)


            # distance = min([bc.distance(center[0]), a.distance(center[0])])
            # print(center[0])

            # plt.scatter(bis1.points[0][0], -bis1.points[0][1], color = "black")
            # plt.scatter(bis1.points[1][0], -bis1.points[1][1], color = "black")


            # plt.scatter(ab.points[0][0], -ab.points[0][1], color = "yellow")
            # plt.scatter(ab.points[1][0], -ab.points[1][1], color = "skyblue")
            # plt.scatter(bc.points[0][0], -bc.points[0][1], color = "pink")
            # plt.scatter(bc.points[1][0], -bc.points[1][1], color = "brown")


            # plt.scatter(center[0][0], -center[0][1], color = "red")
            right_radius = 2 * min(float(bc.distance(center[0])), float(a.distance(center[0]))) / dist

            # print(right_radius)
            # plt.grid()
        elif label["value"]["polygonlabels"][0] == "left-poly":
            left_poly = label["value"]["points"]


            ab = sorted(sorted(left_poly, key=lambda x: -x[0])[:2], key=lambda x: -x[1])

            cd = sorted(sorted(left_poly, key=lambda x: x[0])[:2], key=lambda x: x[1])

            bc = [ab[1], cd[0]]

            ad = [ab[0], cd[1]]

            # plt.scatter(ab[0][0], -ab[0][1], color = "blue")
            # plt.scatter(ab[1][0], -ab[1][1], color = "red")
            # plt.scatter(cd[0][0], -cd[0][1], color = "purple")
            # plt.scatter(cd[1][0], -cd[1][1], color = "orange")
            a = Point(ab[0][0], ab[0][1])

            ab = Line2D(Point(ab[0][0], ab[0][1]), Point(ab[1][0], ab[1][1] - 0.5 * dist))
            cd = Line2D(Point(cd[0][0], cd[0][1] - 0.5 * dist), Point(cd[1][0], cd[1][1]))
            bc = Line2D(Point(bc[0][0], bc[0][1] - 0.5 * dist), Point(bc[1][0], bc[1][1] - 0.5 * dist))
            ad = Line2D(Point(ad[0][0], ad[0][1]), Point(ad[1][0], ad[1][1]))

            bis1 = bc.bisectors(ab)[1] 
            bis2 = bc.bisectors(cd)[1]
            center = bis1.intersection(bis2)


            # distance = min([bc.distance(center[0]), a.distance(center[0])])
            # print(center[0])

            # plt.scatter(bis1.points[0][0], -bis1.points[0][1], color = "black")
            # plt.scatter(bis1.points[1][0], -bis1.points[1][1], color = "black")


            # plt.scatter(ab.points[0][0], -ab.points[0][1], color = "yellow")
            # plt.scatter(ab.points[1][0], -ab.points[1][1], color = "skyblue")
            # plt.scatter(bc.points[0][0], -bc.points[0][1], color = "pink")
            # plt.scatter(bc.points[1][0], -bc.points[1][1], color = "brown")


            # plt.scatter(center[0][0], -center[0][1], color = "red")
            left_radius = 2 * min(float(bc.distance(center[0])), float(a.distance(center[0]))) / dist

            # print(right_radius)
            # plt.grid()
    df.loc[i] = [file_name, left_radius, right_radius]


```

```python
df.to_csv("ground-truth.csv")
```
